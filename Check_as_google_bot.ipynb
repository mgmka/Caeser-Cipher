```python
import requests

def get_site_content(url, user_agent):
    response = requests.get(url, headers={"User-Agent": user_agent})
    if response.status_code == 200:
        return response.content
    else:
        return None

print(get_site_content("https://briefed.online/", "Mozilla/5.0 (compatible; Googlebot/2.1; +http://www.google.com/bot.html)"))
```
